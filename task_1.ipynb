```python
import pandas as pd

file_path = "/content/Inconsistent_Data.xlsx"   # for Colab upload
df = pd.read_excel(file_path)

print("Original Shape:", df.shape)
print(df.head())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print(df.head())
```

```python
df["category"].replace("Unknown", "Accommodation", inplace=True)

df["category"].fillna("Accommodation", inplace=True)

print(df.head())
```

```python
# Removing commas
df["amount"] = df["amount"].astype(str).str.replace(",", "")

# Converting to numeric
df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

# Filling missing values with mean
df["amount"].fillna(df["amount"].mean(), inplace=True)
```

```python
print(df["amount"].dtype)
print(df["amount"].isnull().sum())
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")

print(df["date"].head())
```

```python
df.drop_duplicates(inplace=True)

print("Cleaned Shape:", df.shape)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaning completed. File saved as cleaned_dataset.csv")
```
